## Logistic Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets as datasets
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import PolynomialFeatures

centers = [[0, 0]]
t, _ = datasets.make_blobs(n_samples=750, centers=centers, cluster_std=1, random_state=0)

# LINE
def generate_line_data():
    # create some space between the classes
    X = np.array(list(filter(lambda x : x[0] - x[1] < -.5 or x[0] - x[1] > .5, t)))
    Y = np.array([1 if x[0] - x[1] >= 0 else 0 for x in X])
    return X, Y

# CIRCLE
def generate_circle_data(t):
    # create some space between the classes
    X = np.array(list(filter(lambda x : (x[0] - centers[0][0])**2 + (x[1] - centers[0][1])**2 < 1 or (x[0] - centers[0][0])**2 + (x[1] - centers[0][1])**2 > 1.5, t)))
    Y = np.array([1 if (x[0] - centers[0][0])**2 + (x[1] - centers[0][1])**2 >= 1 else 0 for x in X])
    return X, Y

# XOR
def generate_xor_data():
    X = np.array([
        [0,0],
        [0,1],
        [1,0],
        [1,1]])
    Y = np.array([x[0]^x[1] for x in X])
    return X, Y
```

a) Using the above code, generate and plot data that is linearly separable.

```python
# Generate linearly separable data
X_line, Y_line = generate_line_data()

# Plot the data
plt.scatter(X_line[:, 0], X_line[:, 1], c=Y_line, cmap=plt.cm.Paired)
plt.title('Linearly Separable Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

b) Fit a logistic regression model to the data a print out the coefficients.

```python

# Fit logistic regression model
logreg = LogisticRegression()
logreg.fit(X_line, Y_line)

# Print coefficients
print("Coefficients:", logreg.coef_)
print("Intercept:", logreg.intercept_)
```

c) Using the coefficients, plot the line through the scatter plot you created in a). (Note: you need to do some math to get the line in the right form)

```python
plt.scatter(X_line[:, 0], X_line[:, 1], c=Y_line, cmap=plt.cm.Paired, label='Data')

# Plot the decision boundary (line)
coef = logreg.coef_
intercept = logreg.intercept_

# Plot the decision boundary (line)
x_values = np.linspace(np.min(X_line[:, 0]), np.max(X_line[:, 0]), 100)
y_values = -(coef[0, 0] / coef[0, 1]) * x_values - (intercept / coef[0, 1])

plt.plot(x_values, y_values, '-r', label='Decision Boundary')

plt.title('Linearly Separable Data with Decision Boundary')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
# Function to generate three concentric circles
def generate_concentric_circles():
    theta = np.linspace(0, 2*np.pi, 100)
    circle1 = np.column_stack((np.cos(theta), np.sin(theta)))
    circle2 = 2 * np.column_stack((np.cos(theta), np.sin(theta)))
    circle3 = 3 * np.column_stack((np.cos(theta), np.sin(theta)))

    X = np.vstack([circle1, circle2, circle3])
    Y = np.hstack([np.zeros(100), np.ones(100), 2 * np.ones(100)])

    return X, Y

# Generate concentric circles data
X_concentric, Y_concentric = generate_concentric_circles()

# Polynomial feature transformation
degree = 2  # Degree of the polynomial features (you can adjust this)
poly = PolynomialFeatures(degree=degree)
X_transformed_concentric = poly.fit_transform(X_concentric)

# Fit logistic regression model
logreg_transformed_concentric = LogisticRegression()
logreg_transformed_concentric.fit(X_transformed_concentric, Y_concentric)

# Plot the original data
plt.scatter(X_concentric[:, 0], X_concentric[:, 1], c=Y_concentric, cmap=plt.cm.Paired, label='Original Data')

# Plot the decision boundary in the transformed space
h = .02  # Step size in the mesh
x_min, x_max = X_concentric[:, 0].min() - 1, X_concentric[:, 0].max() + 1
y_min, y_max = X_concentric[:, 1].min() - 1, X_concentric[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

Z = logreg_transformed_concentric.predict(poly.transform(np.c_[xx.ravel(), yy.ravel()]))

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, cmap=plt.cm.Paired, levels=[0.5, 1.5], linestyles='dashed', linewidths=2)

plt.title('Concentric Circles Data with Decision Boundary (Transformed)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
